```python
## sequencia de fibonacci

def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    fib = [0, 1]
    for i in range(2, n + 1):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib

# usando a funcao
fibonacci(10)
```

```python
# 1º Questão - desenvolvendo a função de binet
import math

def fibonacci_binet(n):
    phi = (1 + math.sqrt(5)) / 2   # primeiro termo da Fn
    psi = (1 - math.sqrt(5)) / 2   # segundo termo da Fn

    # Fn completa
    fibonacci_sequence = [round((phi**i - psi**i) / math.sqrt(5)) for i in range(n)]
    return fibonacci_sequence

# definindo os valores do n para verificar a sequencia de fibonacci
n = 1000
# visualizando a sequência
print(fibonacci_binet(n))
```

```python
# 2º Questão - Desenvolvendo uma função da poisson para obter a probabilidade acumulada
import math
def poisson_cdf(x, lam):
    pac = 0.0
    for k in range(0, x + 1):
        pac += (math.exp(-lam) * (lam ** k)) / math.factorial(k)
    return pac

# valores de x e parametro lambda
x = 1
lam = 3
print(f"P(X ≤ {x}) =", poisson_cdf(x, lam))
```

```python
# 3º Questão - Desenvolvendo uma função da poisson para obter a probabilidade acumulada no ponto x
import math
def poisson_cdf(x, lam):
    pac = 0.0
    while (x >= 0):
        proac = (math.exp(-lam) * (lam ** x)) / math.factorial(x)
        pac = pac + proac
        x = x - 1
    return pac

# valores de x e parametro lambda
x = 2
lam = 3
print(f"P(X = {x}) =", poisson_cdf(x, lam))
```

```python
# 4º A série de Maclaurin permite calcular aproximações numéricas de funções.
# Desenvolva a função que retorne essa aproximação. Utilize um suficientemente grande para o limite superior do somatório.
import math
def serie_maclaur(x, k):
    exp = 0
    for i in range(k):
      exp = exp + (x**i) / math.factorial(i)
    return exp
# valores de x e parametro k
x = 2
k = 10
print(f"Aproximação =", serie_maclaur(x, k))

```
