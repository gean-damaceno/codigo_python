```python
# 1 - criando um for para gerar os numeros aleatorios e extrair a média para plotar o historgrama
import math
import matplotlib.pyplot as plt
import random
import numpy as np


def amostrar_normal(N, mi, va, n):
  populacao = np.random.normal(mi, va, N)
  lista = []
  for i in range(N):
    amostra = random.choices(populacao, k = n)
    media = np.mean(amostra)
    x_linha = np.sqrt(N)*(media - mi)
    lista.append(x_linha)

  return lista

N = 100
mi = 25
var = 2
n = 10
medias = amostrar_normal(N, mi, var, n)
```

```python
plt.hist(medias, bins=10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def uniform(n):
    u = np.random.uniform(0, 1, n)
    x = u**(1/3)
    return x
n = 1000
amostra = uniform(n)
plt.hist(amostra, bins=30, density=True, alpha=0.7)
plt.title('Histogram of Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetro da distribuição exponencial
lambda_param = 1.0
N = 1000

# Gerando amostras da distribuição exponencial
samples = np.random.exponential(1 / lambda_param, N)

# Calculando a função de distribuição acumulada (CDF)
x_vals = np.sort(samples)
cdf_vals = 1 - np.exp(-lambda_param * x_vals)

# Plotando o histograma e a CDF
plt.figure(figsize=(10, 6))

# Histograma das amostras
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Histograma')

# Linha da CDF teórica
plt.plot(x_vals, cdf_vals, label='CDF Teórica', color='r', linewidth=2)

plt.title('Distribuição Exponencial e CDF Teórica')
plt.xlabel('x')
plt.ylabel('Densidade / CDF')
plt.legend()
plt.grid(True)
plt.show()
```
